Exploratory Data Analysis

So this is just an basic EDA portfolio ,to increase the understanding regarding various graphs and visualisation tools and libraries involved

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('StudentsPerformance(CSV)')
df.head()
```

```python
##Creating a copy of original DataFrame
df_copy = df.copy()
```

```python
df_copy.columns        ##Finding column names
```

```python
##Finding basic info
df_copy.info()
```

Using Seaborn

Distribution plots

1.Histograms

```python
##Histogram  
sns.histplot(df_copy['gender'] )
plt.title('Gender vs Frequency') 
plt.xlabel('Gender')
plt.ylabel('Counts') 
plt.tight_layout()
plt.show()
```

2.Kernel Density Estimator plot

```python
sns.kdeplot(df_copy['math score'] , fill = True)     ##kdeplot is only for numeric or datetime or continuous variable 
plt.title("Kernel Density plot for 'math score'")  
plt.xlabel('Math score')  
plt.ylabel('Estimated Probablity Density')   
plt.show()
```

3. Distplot

```python
sns.displot(df_copy['lunch'] )
```

Categorical plot

Categorical plots are used to provide relationship between categorical feature and one or more numerical feature

1.Box Plots

```python
sns.boxplot(x = 'test preparation course' , y = 'math score',hue = 'gender', data = df_copy ,palette='Set2')      ##if x = 'categorical feature' then y= "numerical feature" and vice versa
```

```python
sns.boxplot(x = 'race/ethnicity',y = 'reading score' ,hue = 'lunch', data = df_copy, palette='viridis')
```

```python
df_copy.info()
```

```python
sns.boxplot(x ='parental level of education',y = 'writing score', hue = 'test preparation course', palette='Set2' , data= df_copy) 
plt.xticks(rotation = 45) 
plt.show()
```

2. Violin plot

```python
df_copy.info()
```

```python
sns.violinplot(x = 'gender', y='math score', data = df_copy)
```

```python
sns.violinplot(x = 'gender', y='math score',hue ='lunch' , data = df_copy,palette='viridis')
```

3. Bar plot

```python
sns.barplot(x = 'gender' , y= 'reading score' ,data = df_copy)
```

Observation -> Females have a better 'reading score'

```python
sns.barplot(x = 'gender' , y= 'reading score', hue = 'test preparation course' ,data = df_copy , palette= 'Set2')
```

Students who have completed Test Preparation course have performed well

```python
sns.barplot(x = 'parental level of education', y = 'writing score' , data = df_copy)
plt.xticks(rotation = 45)
plt.show()
```

Parents who have done Master's Degree , their children perform well

```python
sns.barplot(x = 'parental level of education', y = 'writing score', hue = 'lunch' ,palette = 'Set2', data = df_copy) 
plt.xticks(rotation = 45)  
plt.show()
```

4. Count plot

```python
sns.countplot(x = 'lunch' ,palette = 'pastel' ,data = df_copy)            ##Count plot has takes only 'x' and mostly a categorical feature
```

Scatter plot shows relationship between two numerical values

3. 3 D plots

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot lines (optional)
ax.plot(df_copy['reading score'], df_copy['math score'], df_copy['writing score'])

# Plot scatter points
ax.scatter(df_copy['reading score'], df_copy['math score'], df_copy['writing score'])

# Label axes
ax.set_xlabel('Reading Score')
ax.set_ylabel('Math Score')
ax.set_zlabel('Writing Score')

plt.show()
```
